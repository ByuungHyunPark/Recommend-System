# Matrix Factorization

- ___m___명의 사용자와 n개의 아이템을 갖고있고 , 이를 m*n 형태의 벡터R로 표현
- 다 오차함수를 최소화하는 k요인 벡터를 찾는것이다  
<center> R≈PQ </center>
- _R_ : m명의 사용자들의 ___n___개의 아이템에 대한 평점 행렬
- _P_ : m명의 사용자와 ___k___요인에 대한 관계 행렬
- _Q_ : n개의 아이템과 k요인에 대한 관계 행렬

모든 사용자들은 모든 상품에 대해서 평점을 매기지 않고 몇몇개만 평점을 매기므로 , 실제 행렬 R은 대부분 희소행렬(sparse matrix)가 된다.

# SVD (singular value decomposition)
